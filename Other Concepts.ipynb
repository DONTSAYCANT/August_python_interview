```python
def sum_ele(*args):
    sum = 0
    print(args)
    print(type(args))
    for i in args:
        sum = sum + i
    return sum

print(sum_ele(10,20))
print(sum_ele(10,20,30))
print(sum_ele(10,10,20,30,40,50,100,1000))
```

```python
def sum_ele(**kwargs):
    print(kwargs)
    for i,j in kwargs.items():
        print(i)
        print(j)

print(sum_ele(val1=10,val2=20))
print(sum_ele(val1=10,val2=20,val3=30))
print(sum_ele(val1=10,val2=10,val3=20,val4=30,val5=40,val6=50,val7=100,val8=1000))
```

```python
from threading import Thread

def func1():
    for i in range(100):
        print("gokul")

        
def func2():
    for i in range(100):
        print("Sarav")
        
def func3():
    for i in range(100):
        print("Ravi")
        

thread1 = Thread(target=func1)
thread2 = Thread(target=func2)
thread3 = Thread(target=func3)

thread1.start()
thread2.start()
thread3.start()
```

```python
from threading import Thread

def func1():
    for i in range(100):
        print("gokul")

        
def func2():
    for i in range(100):
        print("Sarav")
        
def func3():
    for i in range(100):
        print("Ravi")
        

thread1 = Thread(target=func1)
thread2 = Thread(target=func2)
thread3 = Thread(target=func3)

thread1.start()
thread2.start()

thread1.join()
thread2.join()


thread3.start()
```
